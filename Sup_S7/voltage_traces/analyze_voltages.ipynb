# Plot potential of WG4 cells

```python
# Show the particular voltage

f = h5py.File('Data/WG4_stim/cell_vars.h5', 'r')
A=np.array(f['v/data'])

# element 125
voltage=A[:,13]
time=np.linspace(0,160000,16000/0.1)/10.0

plt.plot(time,voltage,c='red')
plt.title('WG4 stim')
plt.xlim((1800, 3200))
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (mV)')
#plt.savefig('GC_stim.pdf',format='pdf')
```
