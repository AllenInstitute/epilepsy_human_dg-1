# Plot potential of WG4 cells

```python

# find cell IDs
cell_ids=[41, 83, 125, 167, 209, 251, 273, 335, 377, 419, 461, 499]
plot_potential(cell_vars_h5='/Data/WG4_stim/cell_vars.h5', gids=[125])
```

```python
# Show the particular voltage

f = h5py.File('WG4_stim/cell_vars.h5', 'r')
A=np.array(f['v/data'])

# element 125
voltage=A[:,13]

plt.plot(voltage)
plt.xlim((18000, 32000))
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (mV)')
plt.savefig('GC_stim.pdf',format='pdf')
```
